```python
dados2015 = pd.read_csv("http://www.anp.gov.br/arquivos/dadosabertos/vc/vendas-gasolinaaviacao-municipio-2015.csv", encoding='iso-8859-1', sep = ';')
dados2015.head()
```

```python
#Comando concat em todos os arquivos usados
dadostotal = pd.concat([dados2013, dados2014, dados2015], ignore_index=True)
dadostotal
```

```python
#Salvando arquivo em formato JSON
dadosJson = dadostotal.to_csv("novo_dados.json", index=False)
```

d) JSON ou JavaScript Object Notation, é um formato de leve troca de dados ou informações entre sistemas e de fácil leitura que também possui uma velocidade maior na execução por ser de arquivo de tamanho reduzido. Sua vantagem sobre o formato CSV é que o JSON mantém estruturas que facilitam a retenção de dados relacionados em um único documento e a representaçào de relações complexas; outra vantagem que podemos citar seria o suporte que o JSON dá a lista de objetos, ajudando a prevenir conversões confusas de listas em um modelo de dados relacional.
