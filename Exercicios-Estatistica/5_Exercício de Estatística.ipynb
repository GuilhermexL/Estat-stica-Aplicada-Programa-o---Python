__1. Determine se os dados são qualitativos ou quantitativos. Explique seu raciocínio.__

(a) Alturas de balões de ar quente;7

(b) Capacidades de carga de caminhonetes;8

(c) Cores dos olhos de modelos;9

(d) Números de identidade de estudantes;10

(e) Respostas em uma pesquisa de opinião.13


_Respostas:_

__a)__ Quantitativo, pois as alturas dos balões são medidas numéricas.

__b)__ Quantitativo, por capacidade ser medido em números

__c)__ Qualitativo, pelo motivo que as cores dos olhos são atributos de um indíviduo. 

__d)__ Quantitativo, porque as identificações de identidade dos estudantes serem de medidas númericas. 

__e)__ Qualitativo, porque as respostas de pesquisas são atributos.
